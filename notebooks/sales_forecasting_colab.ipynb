## 1. Bootstrap the project in Colab

```python
REPO_URL = 'https://github.com/muhfajri24/Sales-Forecasting.git'
COLAB_ROOT = Path('/content/Sales-Forecasting')
LOCAL_NOTEBOOK_ROOT = Path.cwd().resolve()

candidate_roots = [LOCAL_NOTEBOOK_ROOT, LOCAL_NOTEBOOK_ROOT.parent, COLAB_ROOT]
project_root = None
for candidate in candidate_roots:
    if (candidate / 'src' / 'sales_forecasting_pipeline.py').exists():
        project_root = candidate
        break

if project_root is None:
    project_root = COLAB_ROOT
    print('Cloning project repository for Colab runtime...')
    subprocess.run(['git', 'clone', REPO_URL, str(project_root)], check=True)

requirements_file = project_root / 'requirements.txt'
if requirements_file.exists():
    print('Installing project requirements...')
    subprocess.run([sys.executable, '-m', 'pip', 'install', '-q', '-r', str(requirements_file)], check=True)

if str(project_root) not in sys.path:
    sys.path.append(str(project_root))

from src.sales_forecasting_pipeline import run_pipeline
```
